```python
#import statements
import pandas as pd
import json
```

```python
#Review.json  - Properties taken : user_id, business_id, stars, useful, funny, cool, text
reviewCols = ['user_id', 'business_id', 'stars', 'useful', 'funny', 'cool', 'text', 'date']
data = []

with open('D:\\CBS_MSc_Docs\\Courses\\Sem3\\AIML25\\Project\\Yelp-JSON\\Yelp JSON\\yelp_dataset\\yelp_academic_dataset_review.json', encoding='utf-8') as f:
    for line in f:
        record = json.loads(line)
        # Extract only the desired properties
        data.append({col: record.get(col) for col in reviewCols})

df_review = pd.DataFrame(data)

f.close()
data = []
```

```python
#df_review
```

```python
#Read tip.json - Properties taken : user_id, business_id, text, date

tipCols = ['user_id', 'business_id', 'text', 'date']

with open('D:\\CBS_MSc_Docs\\Courses\\Sem3\\AIML25\\Project\\Yelp-JSON\\Yelp JSON\\yelp_dataset\\yelp_academic_dataset_tip.json', encoding='utf-8') as f:
    for line in f:
        record = json.loads(line)
        # Extract only the desired properties
        data.append({col: record.get(col) for col in tipCols})

df_tip = pd.DataFrame(data)

f.close()
data = []
```

```python
#df_tip
```

```python
df_review = pd.concat([df_review, df_tip], axis=0, ignore_index=True)
df_review
```

```python
#Read Business.json

businessCols_drop = ['address', 'city', 'state', 'postal_code', 'latitude', 'longitude', 'is_open', 'hours','hours.Monday', 'hours.Tuesday', 'hours.Wednesday', 'hours.Thursday', 'hours.Friday', 'hours.Saturday', 'hours.Sunday', 'attributes.AcceptsInsurance', 'attributes.HairSpecializesIn']

with open('D:\\CBS_MSc_Docs\\Courses\\Sem3\\AIML25\\Project\\Yelp-JSON\\Yelp JSON\\yelp_dataset\\yelp_academic_dataset_business.json', encoding='utf-8') as f:
    for line in f:
        data.append(json.loads(line))

# Flatten the JSON
df_business = pd.json_normalize(data)

#Remove columns not required
df_business = df_business.drop(businessCols_drop, axis=1)

f.close()
data = []

#df_business.columns
```
